# Instructions
Below you are given 4 lists. One for products, price, product category, and rating. You are hired by the store that owns this data to do some organizing/analysis for them so that they can have a better understanding of their products

````python
import random
categories = ['produce', 'electronics', 'home & kitchen', 'office supplies', 'gardening', 'games', 'beauty', 'appliances', 'movies', 'sports' ]

product_name = [f"product_{num + 1}" for num in range(100)]
product_price = [round(random.uniform(15, 70), 2) for _ in range(100)]
product_category = [random.choice(categories) for _ in range(100)]
product_rating = [round(random.uniform(0, 5), 1) for _ in range(100)]
````

# Objective 1
convert the 4 lists into a list of procduct dictionaries such that each element in the list is a dictionary with the information for each product.

## Example output

````python
#TODO create list of dictionaries here
procucts =[]
for i in range(len(product_name)):
  dictionary = {
      "product":product_name[i],
      "price":product_price[i],
      "category":product_category[i],
      "rating":product_rating[i]
  }
  procucts.append(dictionary)
procucts
````

# Objective 2
Find the average product price and average product rating.

## Example output
```python 
average_product_price = 16.72
average_product_rating = 3.75
```

````python
#TODO find the average price and rating of all products
price_sum = 0
rating_sum = 0
for i in procucts:
  price_sum += i["price"]
  rating_sum += i["rating"]
average_product_price = price_sum / len(procucts)
average_product_sum =   rating_sum / len(procucts)
print(average_product_price, average_product_sum)
````

````python
#TODO create stats dictionary here
group_stats = {}
for i in groups:

  group_stats.setdefault(i,{})
  cost_sum = 0
  highest_cost = 0
  highest_cost_p = ""
  highest_rating = 0
  highest_rating_p = ""
  lowest_rating = 100000000000
  lowest_rating_p = ""
  

  for x in groups[i]:
    cost_sum += x["price"]

    if highest_cost < x["price"]:
      highest_cost = x["price"]
      highest_cost_p = x["product"]
    if highest_rating < x["rating"]:
      highest_rating = x["rating"]
      highest_rating_p = x["product"]
    if lowest_rating > x["rating"]:
      lowest_rating = x["rating"]
      lowest_rating_p = x["product"]

  group_stats[i] = {
      "Average product cost": cost_sum / len(groups[i]),
      "Most expensive product": highest_cost_p,
      "Highest rated product": highest_rating_p,
      "Lowest rated product": lowest_rating_p
  }
  

group_stats
````
