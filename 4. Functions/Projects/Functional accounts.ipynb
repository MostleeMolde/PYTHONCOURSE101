This is part 4 of your bank account project and today you will be moving all of the commands into their own functions.

## Expectations
- Create a dictionary with only one key called `checkings` with a default value of `$50`
- Allow the users to withdraw and deposit money into their `checkings` account 
- Let the users create new accounts with default values of `$0`
- before the user makes a transaction they need to specify which account they're making the transaction in.
- allow users to delete accounts and transfer any remaining amounts to other accounts.
- allow users to transfer money between accounts.
- allow users to see all accounts above a threshold balance

### Commands
* `withdraw` - remove money from account
* `deposit` - add money to account
* `add account` - add a new account to the users bank (allow users to create multiple accounts at once)
* `remove account` - remove an existing account
* `info` - display all accounts and their balances
* `transfer` - move money from a selected account to a target account
* `filter` - given a number, show all accounts with a balance equal to or greater than the number

````python
#TODO
dic = {
    "checking":50
}
def withdraw():
  account = input("What account would you like to withdraw from?")
  value = int(input("Amount?"))
  dic[account] -= value
  print(f"Your new {account} value is {dic[account]}")

def deposit():
  account = input("What account would you like to deposit into?")
  value = int(input("Amount?"))
  print(f"Your new {account} value is {dic[account]}")

def add_account():
  num = int(input("How many accounts do you want to create?"))
  for i in range(num):
    account = input("What's your new account called?")
    value = int(input("How much do you want to deposit into this account?"))
    dic[account] = value

def remove_account():
  account = input("What account are you deleting?")
  trans = input("Transfer money to what account?")
  dic[trans] += dic[account]
  del dic[account]

def info():
  print(dic)
def transfer():
  account = input("From what account?")
  new = input("To what account?")
  value = int(input("How much?"))
  dic[account] -= value
  dic[new] += value

def fillter():
  num = int(input("Number?"))
  for i in dic:
    if dic[i] > num:
      print(dic[i],i) 


functions = {
    "withdraw":withdraw,
    "deposit":deposit,
    "add account":add_account,
    "remove account":remove_account,
    "info":info,
    "transfer":transfer,
    "filter":fillter,
    
}

while True:
  functions[input("What Is Your Input?")]()
````
